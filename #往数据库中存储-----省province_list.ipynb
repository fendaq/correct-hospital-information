# 在county_list表中插入  除了四个直辖市外的所有市   的县

```python
import  pandas as pd
import numpy as np

import pymysql

conn = pymysql.connect(host='127.0.0.1', port=3306, user='root', 
                       passwd='', db='jiadesen', charset='utf8')

cursor = conn.cursor()

for province in province_list[4:]:
    #print(province)
    file_name = province + ".xls"
    print(file_name)
    temp = pd.read_excel(file_name,header = None)
    #print(df)
    L_county = np.array(temp)
    print(L_county)
    county_rows = L_county.shape[0]
    county_cols = L_county.shape[1]
    print(county_rows,county_cols)
    
    for ii in range(county_rows):
        sql3 = "SELECT id FROM city_list WHERE city_name = '"+ L_county[ii][0] +"'"
        #print(sql3)
        #print(L_county[ii][0])
        cursor.execute(sql3)
        #print(cursor.fetchone()[0])
        city_id = str(cursor.fetchone()[0])
        #print(city_id)
        for jj in range(1,county_cols):
            if str(L_county[ii][jj])!= "nan":
                #print(L_county[ii][jj])
                insert_county_sql = "INSERT INTO county_list(city_id, county_name) VALUES(" + city_id + ",'" + L_county[ii][jj] + "')"
                print(insert_county_sql)
                cursor.execute(insert_county_sql)
                conn.commit()
    
cursor.close()
conn.close()           
     
```
